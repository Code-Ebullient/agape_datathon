```python
# Exploration of Women World Cup Dataset

# Step 1 - Importing all the necessary libraries 

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Reading the dataset
wp = pd.read_csv('/Users/concept/Documents/womens-world-cup.csv')
```

```python
# Step 2 - Understanding / getting to know the data 
#   We'll look at the shape | Head/Tail | dtypes | Info | Describe | Columns (series)

# Shape tells us the number of records ie the Rows & Columns
wp.shape
```

```python
# Head/Tail inspects the dataset giving you 5 rows by default
wp.head(20)
```

```python
# dtypes tells us the data types of all the columns. 
# It can be seen as a subset of info() which shows us all the information about the dataset.
wp.info()
```

```python
# Describe, summarizes the dataset statistical properties like mean, min, max etc
wp.describe()
```

```python
# Columns, shows us the list of columns in the dataset. 
# This enables us to study the header and ascertain the necessary ones to use
wp.columns
```

```python
# Step 3 - Data prepation & cleaning
#    We'll identify the duplicates | Drop irrelevant cols & rows | Rename cols | Feature cols

wp.duplicated() # The dataset contains almost int & float, so no duplicate
```

```python
# In the dataset, the most import series is 'goals' 
# so, the series that should be import to goals will be selected
# then, the others drop

wp = wp[['squad', 'year', 'players','matches_played',
       'starts', 'min_playing_time', 'goals', 'assists',
       'non_penalty_goals']].copy()
```

```python
wp.rename(columns= {'squad':'Country',
                    'year':'Year',
                    'players':'Players',
                    'goals':'Tot_goals',
                    'matches_played':'Played',
                    'non_penalty_goals':'Scored',
                    'starts':'Starts',
                    'min_playing_time':'Min_duration',
                    'assists':'Assists'}, inplace = True)
```

```python
# Step 4 - Understanding & Analysing the Feature (Univariate)
#   Using, plotting features (Histogram | kde | Boxplot etc)

wp.columns
```

```python
wp['Tot_goals'].value_counts() \
    .head(10) \
    .plot(kind='barh', title='Total number of goals')
plt.show()
```

```python
wp['Tot_goals'].value_counts() \
    .plot(kind='hist',bins = 30, title='Total number of goals')
plt.show()
```

```python
wp['Scored'].value_counts() \
    .plot(kind='box', title='Total non penalty scored')
plt.show()
```

```python
# Step 5 - Feature relationships (How does these features relate to each other)
#   We used scatterplot | Heatmap correlation | Pairplot | GroupedBy comparism

# Comparism on series x series
wp.plot(kind= 'scatter',
       x= 'Tot_goals',
       y= 'Scored',
       title= 'Total non penalty score vs Total goals')
plt.show()
```

```python
# Using seaborn function on Two features - Total non penalty score vs Total goals
sns.scatterplot(x= 'Tot_goals',
                y= 'Scored',
                #hue= 'Country',
               data= wp)
plt.show()
```

```python
# Comparism on multiple features
sns.pairplot(wp, vars= ['Played', 'Starts', 'Min_duration',
       'Tot_goals', 'Assists', 'Scored'])
```

```python
# Let's reset the index 
# Then ensure that none of the selected features has null values

wp[['Played', 'Starts', 'Min_duration',
       'Tot_goals', 'Assists', 'Scored']] \
    .reset_index(drop= True) \
    .dropna().corr()
```

```python
# Creating the heatmap

wp_corr = wp[['Played', 'Starts', 'Min_duration',
       'Tot_goals', 'Assists', 'Scored']] \
    .reset_index(drop= True) \
    .dropna().corr()
sns.heatmap(wp_corr, annot= True)
```
